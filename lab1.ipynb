```python
import numpy as np
list1=[1,2,3,4,5]
list2=[10,20,30,40,50]

print(type(list1))
print(type(list2))

array1=np.array([1,2,3,4,5])
array2=np.array([10,20,30,40,50])

print(type(array1))
print(type(array2))

multiplication=array1*array2
division=array1/array2
power=array1**2

print("Multiplication:\n",multiplication)
print("Division:\n",division)
print("Power:\n",power)

result_addition=array1*array2
print(f"Addition of Two: \n(array1)+ (array2) (result_addition)")

array1_positive = np.array([1, 2, 3, 4, 5])
sin_values= np.sin(array1)
log_values= np.log(array1_positive)
log2_values= np.log2(array1_positive)
exp_values= np.exp(array1)

print("Sine Values:\n", sin_values)
print("Natural Logarithm Values:\n", log_values)
print("Base-2 Logarithm Values:\n", log2_values)
print("Exponential Values:\n", exp_values)
import matplotlib.pyplot as plt
people= ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40]
height=[145, 151, 165, 173]
weight= [45, 55, 65, 75]

#scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(height, weight, color='blue', s=100)
plt.title('Scatter Plot of Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()

#bar chart
plt.figure(figsize=(10, 6))
plt.bar(people, age, color="green")
plt.title('Bar Chart of Age by Person')
plt.xlabel('Person')
plt.ylabel('Age (years)')
plt.grid(True)
plt.show()

#histogram
plt.figure(figsize=(10, 6))
plt.hist(weight, bins=4, color='purple', edgecolor="black")
plt.title('Histogram of Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
